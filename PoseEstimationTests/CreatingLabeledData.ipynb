```python
# SOURCE: https://www.youtube.com/watch?v=brwgBf6VB0I 

import cv2
import mediapipe as mp
import numpy as np
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()
# to be used to create a fake ground truth(key points actual x,y coordinates) to train our own model
```

```python
def extract_keypoints(results):
    lms = str(results.pose_landmarks).split("}\nlandmark {")

    lms[0] = lms[0][10:]
    lms[-1] = lms[-1][0:-2]
    kps = []

    for i in lms:
        row = i.split("\n")[1:-1]
        floatrow = []
        floatrow.append(float(row[0][5:]))
        floatrow.append(float(row[1][5:]))
        floatrow.append(float(row[2][5:]))
        floatrow.append(float(row[3][14:]))
        kps.append(floatrow)
    return kps
```

```python
# imagine this is image is one we want to train on
# in the preprocessing is where this would all happen

# read in convert from BGR to RGB for the mediapipe model
img = cv2.imread('Custom Dataset/img2.jpg')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
results = pose.process(imgRGB)

# convert the image to grayscale since thats what our model will use when 
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# this method extracts the keypoints from its janky string format to neat 2D float array
keypoints = extract_keypoints(results)
```
